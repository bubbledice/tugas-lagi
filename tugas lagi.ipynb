```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the workflow steps for Change Management
steps_change = [
    "Permintaan Perubahan",
    "Analisis Dampak",
    "Persetujuan Perubahan",
    "Perencanaan & Implementasi",
    "Evaluasi & Penutupan"
]

# Create a directed graph for Change Management
G_change = nx.DiGraph()

# Add edges based on the steps
for i in range(len(steps_change) - 1):
    G_change.add_edge(steps_change[i], steps_change[i + 1])

# Define positions for the nodes to make it linear
pos_change = {step: (0, -i) for i, step in enumerate(steps_change)}

# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(
    G_change, 
    pos_change, 
    with_labels=True, 
    node_size=3000, 
    node_color="lightcoral", 
    font_size=10, 
    font_weight="bold", 
    arrowsize=20
)
plt.title("Change Management Workflow", fontsize=14)
plt.show()
```
